```python
time_data_all = pd.DataFrame()

# 시각화할 행정동 코드 리스트 설정
dong_codes = ["논현2동", "압구정동", "청담동"]  # 필요한 만큼 추가

# 선택한 행정동 코드에 해당하는 데이터 필터링 및 시간대 별 유동인구 수 추출
for dong_code in dong_codes:
    # 특정 행정동에 해당하는 데이터 필터링
    target_dong_data = df[df["행정동_코드_명"] == dong_code]
    
    # 시간대 별 유동인구 수 추출 및 행 추가
    time_columns = ["시간대_00_06_유동인구_수", "시간대_06_11_유동인구_수", "시간대_11_14_유동인구_수", 
                    "시간대_14_17_유동인구_수", "시간대_17_21_유동인구_수", "시간대_21_24_유동인구_수"]
    time_data = target_dong_data[time_columns].sum()
    time_data_all[dong_code] = time_data

# 그래프 그리기
plt.figure(figsize=(10, 6))
time_data_all.plot(kind='bar', ax=plt.gca(), color=['skyblue', 'orange', 'green'])
plt.title('여러 행정동의 시간대 별 유동인구 수')
plt.xlabel('시간대')
plt.ylabel('유동인구 수')
plt.xticks(rotation=45)
plt.legend(title='행정동 코드')
plt.grid(axis='y')
plt.show()
```

```python
df['기준_년분기_코드']=df['기준_년분기_코드'].astype(str)

# 필터링할 특정 행정동 코드 리스트 설정
dong_codes = ["논현2동", "압구정동", "청담동"]  # 필요한 만큼 추가

# 기준 년분기별로 서브플롯을 생성하여 각 행정동의 시간대별 유동인구 수 시각화
fig, axes = plt.subplots(len(dong_codes), 1, figsize=(20, 6 * len(dong_codes)), sharex=True)

for i, dong_code in enumerate(dong_codes):
    # 선택한 행정동 코드에 해당하는 데이터 필터링
    target_dong_data = df[df["행정동_코드_명"] == dong_code]
    
    # 시간대 별 유동인구 수 계산
    time_columns = ["시간대_00_06_유동인구_수", "시간대_06_11_유동인구_수", "시간대_11_14_유동인구_수", 
                    "시간대_14_17_유동인구_수", "시간대_17_21_유동인구_수", "시간대_21_24_유동인구_수"]
    time_data = target_dong_data.groupby("기준_년분기_코드")[time_columns].sum()
    
    # 시간대 별 유동인구 수 시각화
    time_data.plot(ax=axes[i], marker='o', linestyle='-')
    axes[i].set_title(f'기준 년분기별 {dong_code}의 시간대 별 유동인구 수')
    axes[i].set_ylabel('유동인구 수')
    axes[i].set_xlabel('기준 년분기')

    # 년도 일부만 표시하고 x 라벨 회전
    axes[i].set_xticks(range(0, len(time_data.index), 2))
    axes[i].set_xticklabels(time_data.index[::2], rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 특정 행정동 코드 설정
target_dong_code = "청담동"

# 선택한 행정동 코드에 해당하는 데이터 필터링
target_dong_data = df[df["행정동_코드_명"] == target_dong_code]

# 남성과 여성의 총 유동인구 수 계산
total_male_population = target_dong_data["남성_유동인구_수"].sum()
total_female_population = target_dong_data["여성_유동인구_수"].sum()

# 원 그래프 시각화
labels = ['남성', '여성']
sizes = [total_male_population, total_female_population]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # 파이 차트에서 남성 부분을 조금 떼어내기 위한 설정

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title(f'{target_dong_code}의 남녀 유동인구 비율')
plt.axis('equal')  # 원의 형태를 유지함
plt.show()
```

```python
import json

# 특정 행정동 코드 설정
target_dong_code = "청담동"

# 선택한 행정동 코드에 해당하는 데이터 필터링
target_dong_data = df[df["행정동_코드_명"] == target_dong_code]

# 시간대 별 유동인구 수 계산
time_columns = ["시간대_00_06_유동인구_수", "시간대_06_11_유동인구_수", "시간대_11_14_유동인구_수", 
                "시간대_14_17_유동인구_수", "시간대_17_21_유동인구_수", "시간대_21_24_유동인구_수"]

# 년분기를 인덱스로 설정하여 시간대 별 유동인구 수를 그룹화
grouped_data = target_dong_data.groupby("기준_년분기_코드")[time_columns].sum()

# JSON으로 변환
graph_data = {
    "target_dong_code": target_dong_code,
    "data": {
        "time_periods": [col[-8:].replace("_", "-") for col in time_columns],
        "quarters": grouped_data.to_dict(orient="index")
    }
}

# JSON 형식으로 저장
with open("graph_data.json", "w") as json_file:
    json.dump(graph_data, json_file, indent=4)
```
