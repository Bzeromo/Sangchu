```python
# 상권 유형별 평균 시가 구하기
average_price_by_type = df.groupby("상권_구분_코드_명")["아파트_평균_시가"].mean().reset_index()

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(average_price_by_type["상권_구분_코드_명"], average_price_by_type["아파트_평균_시가"], color='skyblue')
plt.title('상권 유형별 평균 아파트 시가')
plt.xlabel('상권 유형')
plt.ylabel('평균 아파트 시가')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 특정 분기의 데이터 필터링
target_quarter = 20231
quarterly_data = df[df["기준_년분기_코드"] == target_quarter]

# 상권 유형별 평균 시가 구하기
average_price_by_type = quarterly_data.groupby("상권_구분_코드_명")["아파트_평균_시가"].mean().reset_index()

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(average_price_by_type["상권_구분_코드_명"], average_price_by_type["아파트_평균_시가"], color='skyblue')
plt.title(f'{target_quarter} 상권 유형별 평균 아파트 시가')
plt.xlabel('상권 유형')
plt.ylabel('평균 아파트 시가')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 특정 상권 코드와 특정 분기 코드 설정
target_market_code = 3110009
target_quarters = [20231, 20232, 20233]  # 필요한 만큼 추가

# 특정 상권의 특정 분기들별 데이터 필터링
filtered_data = df[(df["상권_코드"] == target_market_code) & df["기준_년분기_코드"].isin(target_quarters)]

# 아파트 면적 수의 합 구하기
area_columns = ["아파트_면적_66_제곱미터_세대_수", "아파트_면적_99_제곱미터_세대_수", "아파트_면적_132_제곱미터_세대_수", "아파트_면적_165_제곱미터_세대_수"]
filtered_data["아파트_면적_수_합"] = filtered_data[area_columns].sum(axis=1)

# 각 분기별 아파트 면적 수의 합 그래프로 시각화
plt.figure(figsize=(10, 6))
plt.plot(filtered_data["기준_년분기_코드"], filtered_data["아파트_면적_수_합"], marker='o', linestyle='-')
plt.title(f'{target_market_code} 상권의 특정 분기들별 아파트 면적 수의 합')
plt.xlabel('분기')
plt.ylabel('아파트 면적 수 합')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 특정 상권 코드 리스트 설정
market_codes = [3110009, 3110020, 3110001]  # 필요한 만큼 추가

# 특정 분기 리스트 설정
target_quarters = [20231]  # 필요한 만큼 추가

plt.figure(figsize=(12, 8))

# 각 상권 코드별로 데이터 추출하고 그래프에 추가
for market_code in market_codes:
    for target_quarter in target_quarters:
        # 특정 상권 코드와 특정 분기의 데이터 필터링
        filtered_data = df[(df["상권_코드"] == market_code) & (df["기준_년분기_코드"] == target_quarter)]
        
        # 아파트 면적별 세대 수 추출
        area_columns = ["아파트_면적_66_제곱미터_세대_수", "아파트_면적_99_제곱미터_세대_수", "아파트_면적_132_제곱미터_세대_수", "아파트_면적_165_제곱미터_세대_수"]
        area_data = filtered_data[area_columns].sum()
        
        # 그래프에 추가
        plt.plot(area_data.index, area_data.values, marker='o', label=f'{market_code}, {target_quarter}')

plt.title('상권별 특정 분기들의 아파트 면적별 세대 수')
plt.xlabel('아파트 면적')
plt.ylabel('세대 수')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
